```python
N = input("N:")
x = int(input("x:"))
y = int(input("y:"))

if len(str(x)) != len(str(y)):
    print("NO same lens")
else:
    x_digits = [int(d) for d in str(x)]
    y_digits = [int(d) for d in str(y)]
    for i in x_digits:
        if i in y_digits:
            x_digits.remove(i)
            y_digits.remove(i)
    print(x_digits,"\t",y_digits)
```
